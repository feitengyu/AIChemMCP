```python
import json
import requests
```

```python
url = "http://114.214.215.131:8818/worker/workstation/definition/list?page=1&size=300&release=1"
headers = {
    "identifies": "691c9f24af764bd6ac955a0e8dd0dba9"
}
# 发送GET请求:所有工作站信息
response = requests.get(url, params={}, headers=headers)
result = response.json()
```

```python
for data in result['data']:
    id = data['id']
    try:
        url = f"http://114.214.215.131:8818/worker/workstation/definition/detail?id={id}"
        response = requests.get(url, params={}, headers=headers)
        result_workstation = response.json()['data']
        noteCn = result_workstation['noteCn']
        if len(noteCn) > 0:
            with open(f"./workstation_data/{data['id']}.json", 'w') as f:
                json.dump(result_workstation, f)
            print(result_workstation[''])
            break
    except:
        print("Wrong id: ", data['id'])
```
